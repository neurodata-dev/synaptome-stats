```r
require(mclust)
source("kmpp.r")
data(iris)

x <- iris[,1:4]
spec <- factor(iris$Species); levels(spec) <- c(1:3)
```

```r
ai <- c()
for( i in 1:1e3){
    set.seed(i)
    kpp <- kmpp(x,k=3)
    
    set.seed(i)
    km <- kmeans(x,centers=3)
    
    ari <- mclust::adjustedRandIndex(kpp$cluster, km$cluster)
    ai <- c(ai,ari)
}
```

```r
hist(ai, main="ARI")
plot(x,col=spec, main="TRUTH")
```

```r
plot(x, col=km$cluster, main="K-means")
plot(x, col=kpp$cluster, main="K-means++")
```
